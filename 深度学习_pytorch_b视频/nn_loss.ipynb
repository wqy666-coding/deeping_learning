```python
import torch
from torch.nn import L1Loss
from torch import nn
import torchvision
from torch.utils.data import DataLoader
```

```python
input = torch.tensor([1,2,3],dtype=torch.float32)
target = torch.tensor([1,2,5],dtype=torch.float32)
# 方法一
# input = torch.reshape(input,(1,1,1,3))
# target = torch.reshape(target,(1,1,1,3))
# 方法二
input = input.view(1,1,1,-1)
target = target.view(1,1,1,-1)


# 计算误差（l1）
loss = L1Loss(reduction='sum')# 默认的是求平均
result = loss(input,target)
print(result)


```

```python
# l2，均方误差(回归问题）
loss  = nn.MSELoss()
result = loss(input,target)
print(result)
```

```python
# 交叉熵（分类问题）
x = torch.tensor([0.1,0.2,0.3])
y = torch.tensor([1])
x = torch.reshape(x,(1,3))
loss = nn.CrossEntropyLoss()
result = loss(x,y)
print(result)
```

```python
# 运用
dataset = torchvision.datasets.CIFAR10("./dataset",train=False,transform=torchvision.transforms.ToTensor(),download=False)
dataloader = DataLoader(dataset,batch_size=1)
```

```python
class Tutui2(nn.Module):
    def __init__(self):
        super(Tutui2,self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(3,32,5,padding=2),
            nn.MaxPool2d(2),
            nn.Conv2d(32,32,5,padding=2),
            nn.MaxPool2d(2),
            nn.Conv2d(32,64,5,padding=2),
            nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Linear(1024,64),
            nn.Linear(64,10)
        )
    def forward(self,x):
        x = self.model(x)
        return x
```

```python
loss = nn.CrossEntropyLoss()
tutui = Tutui2()
# 设置优化器(随机梯度下降)
'''
优化器的作用包括：2
参数更新：优化器根据损失函数的梯度信息来更新模型的参数。
加速收敛：优化器可以帮助模型更快地收敛到最优解。
避免局部极小值：优化器能够帮助模型跳出局部极小值，以寻找全局最优解。
处理非凸函数：在深度学习中，损失函数通常是高度非凸的，这意味着存在多个局部最优解。
自适应学习率：一些优化器（如Adam）可以根据参数的梯度自适应地调整学习率，从而在不同参数和时间步骤上使用不同的学习率，提高了优化的稳定性和适应性。

'''
optim = torch.optim.SGD(tutui.parameters(),lr=0.01)
for data in dataloader:
    img , target = data
    output = tutui(img)
    result = loss(output,target)
    # 把可调节参数的梯度调为0
    optim.zero_grad()
    
```
